```python
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# load as Pandas dataframe
json_data=pd.read_json('world_bank_projects.json')
json_data.head(2)
```

# JSON exercise
Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,

Find the 10 countries with most projects

Find the top 10 major project themes (using column 'mjtheme_namecode')

In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
json_data[["countryshortname","project_name"]]
```

```python
data=json_data.groupby(["countryshortname"]).count()[['project_name']].sort_values(["project_name"],ascending=False).rename({'countryshortname': 'TotalProjectName'}, axis=1)
```

```python
#Find the 10 countries with most projects
data[:10]
```

```python
# load json as string
json_load=json.load((open('world_bank_projects.json')))

#Find the top 10 major project themes (using column 'mjtheme_namecode')

project_theme = json_normalize (data=json_load,record_path='mjtheme_namecode',meta=['countryshortname','id'])

project_theme.info()
```

```python
top_10_project_theme=project_theme.groupby(["name"]).count()[["code"]].sort_values(["code"],ascending=False).rename({'name': 'TotalProjects'},axis=1)
top_10_project_theme[:10]
```

```python
#Create a dataframe with the missing names filled in.
normialized_data_copy=project_theme.copy()
```

```python
normialized_data_copy["name"]=normialized_data_copy["name"].apply(lambda x: "Unknown" if x.strip()=='' else x)
```

```python
normialized_data_copy.head(10)
```

```python
normialized_data_copy.groupby(["name"]).count()[["code"]].sort_values(["code"],ascending=False).rename({'name': 'TotalProjects'},axis=1)
```
